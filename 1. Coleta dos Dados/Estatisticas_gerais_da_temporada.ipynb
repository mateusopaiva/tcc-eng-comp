```python
import pandas as pd
from nba_api.stats.endpoints import leaguestandings
from nba_api.stats.endpoints import leaguedashteamstats

team_to_conference = {
    'Atlanta Hawks': 'East', 'Boston Celtics': 'East', 'Brooklyn Nets': 'East',
    'Charlotte Hornets': 'East', 'Chicago Bulls': 'East', 'Cleveland Cavaliers': 'East',
    'Detroit Pistons': 'East', 'Indiana Pacers': 'East', 'Miami Heat': 'East',
    'Milwaukee Bucks': 'East', 'New York Knicks': 'East', 'Orlando Magic': 'East',
    'Philadelphia 76ers': 'East', 'Toronto Raptors': 'East', 'Washington Wizards': 'East',
    'Dallas Mavericks': 'West', 'Denver Nuggets': 'West', 'Golden State Warriors': 'West',
    'Houston Rockets': 'West', 'Los Angeles Clippers': 'West', 'Los Angeles Lakers': 'West',
    'Memphis Grizzlies': 'West', 'Minnesota Timberwolves': 'West', 'New Orleans Pelicans': 'West',
    'Oklahoma City Thunder': 'West', 'Phoenix Suns': 'West', 'Portland Trail Blazers': 'West',
    'Sacramento Kings': 'West', 'San Antonio Spurs': 'West', 'Utah Jazz': 'West'
}

# Lista de temporadas
seasons = ['2001-02', '2002-03', '2003-04', '2004-05', '2005-06', '2006-07', '2007-08',
           '2008-09', '2009-10', '2010-11', '2011-12', '2012-13', '2013-14', '2014-15',
           '2015-16', '2016-17', '2017-18', '2018-19', '2019-20', '2020-21', '2021-22',
           '2022-23', '2023-24']

def get_team_stats_for_season(season):
    season = season
    params = {
        'season': str(season) + '-' + str(season + 1)[-2:],
        'season_type_all_star': 'Regular Season'
    }
    team_stats = leaguedashteamstats.LeagueDashTeamStats(**params)
    team_stats_df = team_stats.get_data_frames()[0]

    # Renomear nome das equipes
    team_stats_df.replace('Charlotte Bobcats', 'Charlotte Hornets', inplace=True)
    team_stats_df.replace('LA Clippers', 'Los Angeles Clippers', inplace=True)
    team_stats_df.replace('New Jersey Nets', 'Brooklyn Nets', inplace=True)
    team_stats_df.replace('New Orleans Hornets', 'New Orleans Pelicans', inplace=True)
    team_stats_df.replace('Seattle SuperSonics', 'Oklahoma City Thunder', inplace=True)
    team_stats_df.replace('New Orleans/Oklahoma City Hornets', 'New Orleans Pelicans', inplace=True)
    team_stats_df.replace('Charlotte Hornets', 'Charlotte Hornets', inplace=True)

    # Adicionando a coluna 'CONFERENCE' com base no mapeamento
    team_stats_df['CONFERENCE'] = team_stats_df['TEAM_NAME'].map(team_to_conference)
    team_stats_df['SEASON'] = season
    return team_stats_df

def get_stats_for_multiple_seasons(seasons):
    all_seasons_stats = []
    for season in seasons:
        season_start = int(season.split('-')[0])
        season_stats = get_team_stats_for_season(season_start)
        all_seasons_stats.append(season_stats)

    combined_stats = pd.concat(all_seasons_stats)
    return combined_stats

# Obter estatísticas para todas as temporadas
all_seasons_data = get_stats_for_multiple_seasons(seasons)
```
