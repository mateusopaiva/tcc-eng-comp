```python
from nba_api.stats.endpoints import leaguestandings
import pandas as pd

# Criar um dicionário de mapeamento para substituição
team_name_mapping = {
    'Nets': 'Brooklyn Nets',
    'Kings': 'Sacramento Kings',
    'Pistons': 'Detroit Pistons',
    'Spurs': 'San Antonio Spurs',
    'Celtics': 'Boston Celtics',
    'Lakers': 'Los Angeles Lakers',
    'Hornets': 'Charlotte Hornets',
    'Mavericks': 'Dallas Mavericks',
    'Timberwolves': 'Minnesota Timberwolves',
    'Magic': 'Orlando Magic',
    '76ers': 'Philadelphia 76ers',
    'Trail Blazers': 'Portland Trail Blazers',
    'Raptors': 'Toronto Raptors',
    'SuperSonics': 'Oklahoma City Thunder',
    'Pacers': 'Indiana Pacers',
    'Jazz': 'Utah Jazz',
    'Clippers': 'Los Angeles Clippers',
    'Bucks': 'Milwaukee Bucks',
    'Wizards': 'Washington Wizards',
    'Suns': 'Phoenix Suns',
    'Heat': 'Miami Heat',
    'Rockets': 'Houston Rockets',
    'Hawks': 'Atlanta Hawks',
    'Nuggets': 'Denver Nuggets',
    'Grizzlies': 'Memphis Grizzlies',
    'Knicks': 'New York Knicks',
    'Cavaliers': 'Cleveland Cavaliers',
    'Warriors': 'Golden State Warriors',
    'Bulls': 'Chicago Bulls',
    'Bobcats': 'Charlotte Hornets',
    'Thunder': 'Oklahoma City Thunder',
    'Pelicans': 'New Orleans Pelicans'
}

# Criar um dicionário de mapeamento para substituição de nomes de colunas
column_name_mapping = {
    'SeasonID': 'SEASON_ID',
    'TeamName': 'TEAM_NAME',
    'PlayoffRank': 'PLAYOFF_RANK',
    'Conference': 'CONFERENCE',
}

# Lista de temporadas
seasons = ['2001-02', '2002-03', '2003-04', '2004-05', '2005-06', '2006-07', '2007-08',
           '2008-09', '2009-10', '2010-11', '2011-12', '2012-13', '2013-14', '2014-15',
           '2015-16', '2016-17', '2017-18', '2018-19', '2019-20', '2020-21', '2021-22',
           '2022-23', '2023-24']

# Colunas de interesse
columns_of_interest = ['SeasonID', 'TeamName', 'PlayoffRank', 'Conference']

# Inicializar uma lista vazia para todas as classificações
all_standings = []

# Iterar sobre cada temporada
for season in seasons:
    standings = leaguestandings.LeagueStandings(season=season)
    standings_df = standings.get_data_frames()[0]

    # Filtrar e selecionar colunas de interesse
    season_standings = standings_df[columns_of_interest]

    # Adicionar as classificações à lista
    all_standings.append(season_standings)

# Concatenar os resultados em um único DataFrame
all_standings_df = pd.concat(all_standings)

# Substituir os valores no DataFrame
all_standings_df.replace({'TeamName': team_name_mapping}, inplace=True)
all_standings_df.rename(columns=column_name_mapping, inplace=True)

# Alterando o nome da coluna SEASON_ID para SEASON
all_standings_df.rename(columns={'SEASON_ID': 'SEASON'}, inplace=True)


# Removendo o primeiro caractere '2' da coluna SEASON
all_standings_df['SEASON'] = all_standings_df['SEASON'].astype(str).str[1:]

# Exibir o DataFrame atualizado
all_standings_df
```

```python
unique_team_names = all_standings_df['TEAM_NAME'].unique()

for team_name in unique_team_names:
    print(team_name)
```
